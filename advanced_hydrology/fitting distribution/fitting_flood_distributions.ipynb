```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import lmoments as lm
import scipy.stats as stats
%matplotlib inline
```

```python
mpl.rcParams['xtick.labelsize'] = 15
mpl.rcParams['ytick.labelsize'] = 15
mpl.rcParams['lines.linewidth'] = 2.5
mpl.rcParams['axes.labelsize'] = 18
mpl.rcParams['legend.fontsize'] = 15
```

```python
""" function to read data to a dataframe, Gringorten plotting position is used to calculate empirical probabilities """
def get_df(data):
    df_t = pd.DataFrame(data, columns=['flow'])
    df = df_t.sort(['flow'], ascending=False)
    num = data.size
    df['rank'] = np.arange(1, num + 1)
    df['prob'] = (df['rank'] - 0.44)/(num + 0.12)
    return df
```

```python
""" calculate quantiles of given probabilities for each given distribution; L-moment is used to estimate 
    distribution parameters """
def lm_quantiles(data,method,probs):
    npara = dist_dict['nparams'][method]
    lms = lm.samlmu(data,npara)
    paras = dist_dict['parameter'][method](lms)
    quantiles = dist_dict['quantile'][method](probs,paras)
    return quantiles
```

```python
"""estimate peak floods for given return_periods; method is distribution used"""
def estimate_floods(data,methods,return_periods):
    probs = 1 - 1.0/return_periods
    df = pd.DataFrame(index = return_periods)
    for method in methods:
        npara = dist_dict['nparams'][method]
        lms = lm.samlmu(data,npara)
        paras = dist_dict['parameter'][method](lms)       
        quantiles = dist_dict['quantile'][method](probs,paras)
#        print "{}:".format(method)
        df[method] = np.int_(quantiles)
#        for i in np.arange(return_periods.size):
 #                   print "{} Years: {} ".format(return_periods[i],int(quantiles[i]))
    return df
```

```python
"""plot exceedance probabilities for annual peak flow, a few distributions are fitted to flow data"""
def plot_figures(df,basin,methods,probs):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    ax.semilogy(df['flow'],df['prob'],'ok',label='data')
    for method in methods:
        quantiles = lm_quantiles(df['flow'],method,probs)  
        label = dist_dict['method'][method]
        ax.semilogy(quantiles,1-probs,label=label)
#    flow = log_pearson(df,basin)
 #   ax.semilogy(flow,1-prob_LP3,label='Log Pearson III')
    ax.set_xlabel('Annual Peak Flow (cfs)')
    ax.set_ylabel('Probability of Exceedance')
    ax.legend(loc='upper right')
    ax.set_xlim(left=0)
    ax.set_ylim(bottom=0.001)
```
