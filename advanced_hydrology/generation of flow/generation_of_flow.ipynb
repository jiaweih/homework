```python
year = np.genfromtxt('data.txt',skiprows=1,usecols=0)
year = np.int_(year)
obs_flow = np.genfromtxt('data.txt',skiprows=1,usecols=1)
ts = pd.DataFrame(obs_flow, index=year, columns=['obs_flow'])
```

```python
'''
Estimate statistics of Y (logX)
Input
--------------
meanX: mean of sample
varX: variance of sample
skewX: skewness of sample
acX: lag-1 correlation coefficient of sample
--------------
Output
meanY: mean of Y
varY: variance of Y
acY: lag-1 coefficient of Y
a: calibration parameter of model
'''
def estimate_Y_statistic(meanX,varX,skewX,acX):
    for varY in np.arange(0,1,0.01):
        skew_X = (np.exp(3*varY) - 3*np.exp(varY) + 2)/(np.exp(varY) - 1)**(3.0/2.0)
        if np.abs(skew_X - skewX)<0.01:
            break

    for meanY in np.arange(11.2,11.3,0.00001):
        var_X = np.exp(2*(varY + meanY)) - np.exp(varY + 2*meanY)
        if np.abs(var_X - varX)<1000:
            break

    a = meanX - np.exp(varY/2.0 + meanY)

    for acY in np.arange(0.4,0.43,0.0001):
        ac_X = (np.exp(varY*acY) - 1)/(np.exp(varY) - 1)
        if np.abs(ac_X - acX)<0.00005:
            break
    return (meanY, varY, acY, a)                                                
```

```python
'''
Estimate statistics for autoregressive model.
Input
---------
Years: Number of events
Ymean: mean of estimated Y (logX)
r: lag-1 correlation coefficient
a: calibration parameter of model
---------
Output
---------
meanX: mean of modeled X
stdX: std of modeled X
meanY: mean of modeled Y
stdY: std of modeled Y
skewY: skew of modeled Y
'''
def AR_estimation(Years, Ymean, Ystd, r, a):
    Y = np.empty(Years)
    X = np.empty(Years)
    Ypar=Ystd*(1-r**2)**0.5
    Y[0] = Ymean
    X[0] = np.exp(Y[0])
    for i in np.arange(1,Years):
        Y[i] = Ymean + r*(Y[i-1] - Ymean) + (Ypar*np.random.normal(0,1))
        X[i] = np.exp(Y[i]) + a
        
    meanX = np.mean(X)
    stdX = np.std(X)
    skewX = stats.skew(X)
    
    meanY = np.mean(Y)
    stdY = np.std(Y)
    skewY = stats.skew(Y)
    
    hurst = pyeeg.hurst(X)
    acf = sm.tsa.acf(ts, nlags=Years) 
    
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    ax.plot(np.arange(1,Years+1),X)
    ax.set_xlabel('Generated Years')
    ax.set_ylabel('Annual Flow (cfs)')
    
    return X, meanX, stdX, skewX, meanY, stdY, skewY, hurst, acf[1]
```

#Question b

##Mann-Kendall test

>Null hypothesis, H0: the distribution of annual peak discharge does not change as a function of time.

>Kendall's correlation coefficient for annual mean discharge with time is tau = 0.269, p = 0.0006, Z = 3.417, 
    since |Z| > Z(0.975) = 1.96, p < alpha = 0.05, there is enough evidence to reject H0: tau = 0. The observed trend
    is statistically significant at alpha = 0.05.

```python
tau, p_value = stats.kendalltau(ts.index,ts['obs_flow'].values)
z = stats.norm.ppf(1-p_value/2,0,1)
print 'tau = {}, p_value = {}, Z = {}'.format(tau,p_value,z)
```
