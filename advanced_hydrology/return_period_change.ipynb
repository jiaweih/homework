```python
data_flow = np.genfromtxt('question1_data',skip_header=1)
dates = pd.date_range('1920',periods=89,freq='A')
df_flow = pd.DataFrame(data_flow, index = dates, columns=['Year','Flow'])
```

```python
N = 1970 - 1920 + 1
df_flow_pre1970 = df_flow.ix[0:N].sort('Flow')
df_flow_pre1970['rank'] = np.arange(1,N+1)
df_flow_pre1970['cum_P'] = df_flow_pre1970['rank']*1.0/(N+1)
df_flow_pre1970['exceed_P'] = 1 - df_flow_pre1970['cum_P']
df_flow_pre1970['RP'] = 1/df_flow_pre1970['exceed_P']
```

### Q-RP relationship before 1970

### Estimation of the 5-, 10-, 25-year floods using the data before 1970
* Q_5 = 19430
* Q_10 = 29425
* Q_25 = 46050

```python
fig = plt.figure(figsize = (10,8))
ax = fig.add_subplot(111)
ax.plot(df_flow_pre1970['RP'],df_flow_pre1970['Flow'])
ax.set_xlabel('Return Period (Years)')
ax.set_ylabel('Annual Peak FLow (cfs)')
fig.savefig('Q-PQ_before1970.png',dpi=300)
```

```python
Q_5 = 19430
Q_10 = 29425
Q_25 = 46050
```

```python
N = 2008 - 1920 + 1
df_flow_full = df_flow.sort('Flow')
df_flow_full['rank'] = np.arange(1,N+1)
df_flow_full['cum_P'] = df_flow_full['rank']*1.0/(N+1)
df_flow_full['exceed_P'] = 1 - df_flow_full['cum_P']
df_flow_full['RP'] = 1/df_flow_full['exceed_P']
```

###Q-RP relationship for all time

### Estimation of the 5-, 10-, 25-year floods using the full length of data
* Q_5 = 22900
* Q_10 = 33500
* Q_25 = 38400

```python
fig = plt.figure(figsize = (10,8))
ax = fig.add_subplot(111)
ax.plot(df_flow_pre1970['RP'],df_flow_pre1970['Flow'],label = 'Q-RP (1920 - 1970)')
ax.plot(df_flow_full['RP'],df_flow_full['Flow'],label = 'Q-RP (1920 - 2008)')
ax.set_xlabel('Return Period (Years)')
ax.set_ylabel('Annual Peak FLow (cfs)')
ax.legend(loc='upper right',**{'frameon':False})
ax.set_ylim([0,70000])
fig.savefig('Q-PQ_full.png',dpi=300)
```

**Comparing the Q-RP relationship from 1920 to 1970 and from 1920 to 2008, the relationship for all years indicates the annual
peak flow is higher for small return periods and lower for large return periods.**

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)
ax = df_flow['Flow'].plot(label = 'Peak Flow')
ax.set_ylabel('Annual Peak FLow (cfs)')
ax.axhline(y = Q_5, label = 'Q5', c= 'c', ls = '--')
ax.axhline(y = Q_10, label = 'Q10', c = 'm', ls = '--')
ax.axhline(y = Q_25, label = 'Q25', c = 'r', ls = '--')
ax.legend(loc = 'upper right',**{'frameon':False})
plt.grid(False)
ax.set_ylim([0,70000])
fig.savefig('timeseries.png',dpi=300)
```

##Test the hypothesis for the probability of exceeding Q5

###Null hypothesis, H0: The probability of exceedances have not changed for Q5

**T = 25 (Numer of times Q5 is exceeded). Given P0 = 0.2, and n = 89, **

**for alpha = 0.05, Z(alpha/2) = -1.96, Z(1-alpha/2) = 1.96, we found TL = 10.40,TU = 25.20**

**for alpha = 0.1, Z(alpha/2) = -1.64, Z(1-alpha/2) = 1.64, we found TL = 11.61,TL = 23.99**

**For alpha = 0.05, TL < T = 25 < TU, H0 is accepted.**

**For alpha = 0.1, T = 25 > TL, H0 can be rejected.**

```python
Z_05 = {'alpha/2':-1.96,'1-alpha/2':1.96}
Z_01 = {'alpha/2':-1.64,'1-alpha/2':1.64}
```

```python
df_Q5 = df_flow_full['Flow']>=Q_5
T_Q5 = np.count_nonzero(df_Q5)
P0 = 1.0/5
n = df_flow['Flow'].count()
print 'T_Q5 = ', T_Q5
print 'P0 = ', P0
print 'n = ', n
```

###alpha = 0.05

```python
T_L = n*P0 + Z_05['alpha/2']*np.sqrt(n*P0*(1-P0))
T_U = n*P0 + Z_05['1-alpha/2']*np.sqrt(n*P0*(1-P0))
print T_L,T_U
```

###alpha = 0.1

```python
T_L = n*P0 + Z_01['alpha/2']*np.sqrt(n*P0*(1-P0))
T_U = n*P0 + Z_01['1-alpha/2']*np.sqrt(n*P0*(1-P0))
print T_L,T_U
```
