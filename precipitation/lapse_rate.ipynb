```python
def get_df(ifile,name):
    data = np.genfromtxt(ifile)
    header= ['year','month','a','b','c',name]
    df = pd.DataFrame(data, columns = header)
    return df
```

```python
def get_df_month(ifile):
    data = np.genfromtxt(ifile)
    header= np.arange(1,13)
    df = pd.DataFrame(data, columns = header)
    return df
```

```python
df_dry_grant = get_df('dry_grant.txt','dry_grant')
df_storm_grant = get_df('storm_grant.txt','storm_grant')
df_dry_paradise = get_df('dry_paradise.txt','dry_paradise')
df_storm_paradise = get_df('storm_paradise.txt','storm_paradise')
```

```python
fig = plt.figure(figsize = (10,8))
ax = fig.add_subplot(111)
ax.boxplot([df_dry_grant['dry_grant'],df_storm_grant['storm_grant'],df_dry_paradise['dry_paradise'],
            df_storm_paradise['storm_paradise']])
ax.set_xticklabels(['dry,Grant','storm,Grant','dry,Paradise','storm,Paradise'])
ax.axhline(y=5,color='m',lw=2,label='wet adiabatic lapse rate')
ax.axhline(y=9.8,color='g',lw=2,label='dry adiabatic lapse rate')
ax.set_ylim(0,16)
ax.set_ylabel('Lapse Rate ($^\circ$C/km)')
ax.legend(loc='upper left')
fig.savefig('lapse_rate.png',dpi=300)
```

```python
def dew_point(p,T,w):
    e = w*p/(w+0.622)
    es = 33.8639*(np.power(0.00738*T+0.8072,8)-0.000019*abs(1.8*T+48)+0.001316)
    r = e/es
    x = 1 - r/100
    Temp = (14.55+0.114*T)*x + np.power((2.5+0.007*T)*x,3) + (15.9+0.117*T)*np.power(x,14)
    Td = T - Temp
    return e,es,r,Td
```
