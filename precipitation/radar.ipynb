```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import os
import netCDF4 as nc
%matplotlib inline
```

```python
def plot_ts(data_pnw,data_satellite,ofile):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    ax.plot(np.arange(1,13),data_pnw,'k-',label='Drought monitor')
    ax.plot(np.arange(1,13),data_satellite,'b-',label='Stage 4')
    ax.set_xlabel('Month')
    ax.set_ylabel('Precipitation (inch/day)')
    ax.legend(loc='upper center')
    fig.savefig(ofile,dpi=300)
```

```python
def plot_ts_3(data_pnw,data_level,data_stage,ofile):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    ax.plot(np.arange(1,13),data_pnw,'k-',label='Drought monitor')
    ax.plot(np.arange(1,13),data_level,'b-',label='Level 3')
    ax.plot(np.arange(1,13),data_stage,'r-',label='Stage 4')
    ax.set_xlabel('Month')
    ax.set_ylabel('Precipitation (inch/day)')
    ax.legend(loc='upper center')
    fig.savefig(ofile,dpi=300)
```

```python
def get_prcp(prcp_data,lat_data,lon_data):
    total_length = lat_data.shape[0]*lat_data.shape[1]
    lat_tmp = lat_data.reshape(total_length,1)
    lon_tmp = lon_data.reshape(total_length,1)
    prcp_tmp = prcp_data.reshape(total_length,1)
    count=0
    for i in np.arange(total_length):
        if 0<prcp_tmp[i,0]<10000:
            count = count+1    
    lats = np.random.rand(count,1)
    lons = np.random.rand(count,1)
    prcp = np.random.rand(count,1)
    ii=0
    for i in np.arange(total_length):
        if 0<prcp_tmp[i,0]<100000:
            lats[ii] = lat_tmp[i,0]
            lons[ii] = lon_tmp[i,0]*-1
            prcp[ii] = prcp_tmp[i,0]
            ii = ii +1
    return prcp,lats,lons
```

```python
lat_file = 'latitude.txt'
lon_file = 'longitude.txt'
lat_data = np.genfromtxt(lat_file)
lon_data = np.genfromtxt(lon_file)
```

```python
def get_mean(prcp,lats,lons):
    lat_basin = np.genfromtxt('lat.txt')
    lon_basin = np.genfromtxt('lon.txt')
    index = []
    prcp_basin = []
    size = lats.size
    size_basin = lat_basin.size
    for i in np.arange(size_basin):
        for j in np.arange(size):
            if lats[j,0] == lat_basin[i] and lons[j,0] == lon_basin[i]:
                index.append(j)
                prcp_basin.append(prcp[j])
    prcp_daily_mean = np.sum(prcp_basin)*39.3701/(1000*np.size(prcp_basin))    ## daily mean precipitation for basin (inch/day)
    print prcp_daily_mean
```

```python
'''prcp_data are daily mean precipitation for whole U.S. (kg/m-2)'''

infile = 'daily_mean.nc'
rootgroup = nc.Dataset(infile,'r',format='NETCDF4')
prcp_data = rootgroup.variables['Total_precipitation_surface_24_Hour_Accumulation'][:]
rootgroup.close()

prcp,lats,lons = get_prcp(prcp_data,lat_data,lon_data)
get_mean(prcp,lats,lons)
```
