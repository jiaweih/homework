### Function for plotting shortwave and longwave radiation, which are based on day of year

```python
def plot_yday(df,fig_size,variables,fmts,labels,ld,store_dir):
    fig = plt.figure(figsize = fig_size)
    ax = fig.add_subplot(111)
    for idx,var in enumerate(variables):
        fmt = fmts[idx]
        txt = labels[var]
        df[var].groupby(lambda x:x.timetuple().tm_yday).mean().plot(style=fmt,label=txt,linewidth=ld)
    #df[var].groupby(lambda x:get_hour(x)).mean().plot()
    ax.set_xlim([0.5,366.5])
    ax.set_xlabel('Day of Year',fontsize = 20)
    ax.set_ylabel('Radiation ($W/m^2$)',fontsize = 20)
   # ax.set_xticklabels(np.arange(0,24))
    ax.tick_params(axis='both',which='major',labelsize=18)
    ax.legend(loc = 'upper right',prop={'size':15},**{'frameon':False})
   #store_dir = '{}_diurnal.png'.format(var)
    fig.savefig(store_dir,dpi=300)
```

### Function for plotting variables based on diurnal cycle, i.e. hour of a day

```python
def plot_diurnal(df,fig_size,variables,fmts,labels,ld,store_dir):
    fig = plt.figure(figsize = fig_size)
    ax = fig.add_subplot(111)
    for idx,var in enumerate(variables):
        fmt = fmts[idx]
        txt = labels[var]
        df[var].groupby(lambda x:x.timetuple().tm_hour).mean().plot(style=fmt,label=txt,linewidth=ld)
    #df[var].groupby(lambda x:get_hour(x)).mean().plot()
    ax.set_xlim([-0.5,24.5])
    ax.set_xlabel('Hour',fontsize = 20)
    ax.set_ylabel('Radiation ($W/m^2$)',fontsize = 20)
   # ax.set_xticklabels(np.arange(0,24))
    ax.tick_params(axis='both',which='major',labelsize=18)
    ax.legend(loc = 'upper right',prop={'size':15},**{'frameon':False})
   #store_dir = '{}_diurnal.png'.format(var)
    fig.savefig(store_dir,dpi=300)
```

### Function for plotting variables based on seasonal cycle;i.e.month of year

```python
def plot_seasonal(df,fig_size,variables,fmts,labels,ld,ymin,ymax,store_dir):
    fig = plt.figure(figsize = fig_size)
    ax = fig.add_subplot(111)
    for idx,var in enumerate(variables):
        fmt = fmts[idx]
        txt = labels[var]
        df[var].groupby(lambda x:x.timetuple().tm_mon).mean().plot(style=fmt,label=txt,linewidth=ld)
    #df[var].groupby(lambda x:get_hour(x)).mean().plot()
    ax.set_xlim([0.5,12.5])
    ax.set_xlabel('Month',fontsize = 20)
    ax.set_ylabel('Radiation ($W/m^2$)',fontsize = 20)
   # ax.set_xticklabels(np.arange(0,24))
    ax.tick_params(axis='both',which='major',labelsize=18)
    ax.legend(loc = 'upper right',prop={'size':15},**{'frameon':False})
    ax.set_ylim([ymin,ymax])
   # store_dir = '{}_seasonal.png'.format(var)
    fig.savefig(store_dir,dpi=300)
```

## Comparing observed shortwave radiation and simulated shortwave radiation

### Preparing input file for MTCLIM model; input files include Tmax,Tmin,Prec,Tdew

```python
start_init = dt.datetime(2007,1,1)
end_init = dt.datetime(2012,12,31)
dates_init = pd.date_range(start=start_init,end=end_init,freq='D')
df_init = pd.DataFrame(index=dates_init)
```

```python
df_init['Yearday'] = df_Kd['JD'].resample('D',how='mean')
df_init['Tmax'] = df_Kd['TA'].resample('D',how='max')
df_init['Tmin'] = df_Kd['TA'].resample('D',how='min')
df_init['PREC'] = df_Kd['PREC'].resample('D',how='sum')
df_init['Tdew'] = df_Kd['Tdew'].resample('D',how='mean')
```

```python
df_init['Yearday'] = df_init['Yearday'].map(lambda x: '%5d' % x)
df_init['Tmax'] = df_init['Tmax'].map(lambda x: '%5.2f' % x)
df_init['Tmin'] = df_init['Tmin'].map(lambda x: '%5.2f' % x)
df_init['PREC'] = df_init['PREC'].map(lambda x: '%5.2f' % x)
df_init['Tdew'] = df_init['Tdew'].map(lambda x: '%5.2f' % x)
```

```python
output = '{}.txt'.format(name)

df_init.to_csv(output,sep='\t',index=False,line_terminator='\n')
```

```python

df_Kd['SW'] = df_Kd['Rg'] 
df_Kd['LW'] = df_Kd['Rgl'] 
```

### miss84.mtc43 is the output file of MTCLIM model, which includes simulated shortwave radiation

```python
index = df_init.index
header_r = ['yday','Tmax','Tmin','TA','PREC','VPD','SW_sim','daylen']
data_r = np.genfromtxt('miss84.mtc43',delimiter='',skip_header=4)
df_radiation = pd.DataFrame(data_r,index=index,columns=header_r)
```

```python
df_radiation['SW'] = df_Kd['SW'].resample('D',how='mean')
df_radiation['LW'] = df_Kd['LW'].resample('D',how='mean')
df_radiation['RH'] = df_Kd['RH'].resample('D',how='mean')
```

```python
df_radiation=df_radiation.fillna(method='pad')
```

```python
variables_SW = ['SW','SW_sim']
store_dir = 'SW_diurnal_{}.png'.format(name)
ld = 3
plot_yday(df_radiation,fig_size,variables_SW,fmts,labels,ld,store_dir)
```

### Calcuate simulated longwave radiation based on simulated shortwave radiation; Prata method is adopted

```python
sigma = 5.670373*1E-8
df_radiation['K'] = df_radiation['TA'] + 273.15
df_radiation['saturation_pressure'] = 0.1*np.exp(77.3450 + 0.0057*df_radiation['K'] - 7235/df_radiation['K'])/df_radiation['K']**8.2
df_radiation['vapor_pressure'] = df_radiation['RH'] * df_radiation['saturation_pressure']/100
df_radiation['precipitable'] = 4650*df_radiation['vapor_pressure']/df_radiation['K']
df_radiation['coefficient'] = sigma*df_radiation['K']**4
df_radiation['LW_sim'] = (1 - (1 + df_radiation['precipitable'])*np.exp(-np.power((1.2+3.0*df_radiation['precipitable']),0.5)))*df_radiation['coefficient']
```

```python
variables_LW = ['LW','LW_sim']
store_dir = 'LW_diurnal_{}.png'.format(name)
plot_yday(df_radiation,fig_size,variables_LW,fmts,labels,ld,store_dir)
```
