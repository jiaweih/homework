###average

```python
data = np.genfromtxt('./output/precip.kw.Jiawei.txt')
data.shape
```

```python
prcp = np.zeros((360,26))

for j in np.arange(1,26):
    for i in np.arange(38):
        for k in np.arange(360):
            prcp[k,j] = prcp[k,j] + data[360*i+k,j]

for j in np.arange(1,26):
    for k in np.arange(360):
        prcp[k,j] = prcp[k,j]/38.0
```

```python
nstat = 25
periods = 360
slot = 40
time = data[0:periods,0]

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(time[:slot],prcp[:slot,1],label = 'Station 1')
ax.plot(time[:slot],prcp[:slot,5],label = 'Station 5')
ax.plot(time[:slot],prcp[:slot,9],label = 'Station 9')
ax.plot(time[:slot],prcp[:slot,15],label = 'Station 15')
ax.plot(time[:slot],prcp[:slot,19],label = 'Station 19')
ax.legend(loc='upper right')
ax.set_xlabel('Time (Hours)')
ax.set_ylabel('Intensity (mm/hr)')
fig.savefig('Hyetograph.png',dpi=300)
```

###depth isolines

```python
prcp_sum = np.zeros((36,26))

for j in np.arange(1,26):
    for k in np.arange(36):
        prcp_sum[k,j] = np.sum(prcp[k*10:(k+1)*10,j])
```

```python
depth = np.zeros((36,5,5))
for i in np.arange(36):
    m = 0
    for j in np.arange(4,-1,-1):
        for k in np.arange(0,5):
            depth[i,j,k] = (prcp_sum[i,m])/100.0   ##divided by area
            m += 1
```

###depth for first hour

```python
fig = plt.figure()
ax = fig.add_subplot(111)
cs = ax.contour(np.arange(5),np.arange(5),depth[0,:,:])  ###first hour
plt.clabel(cs,fmt='%1.1f')
#plt.xticks(np.arange(-0.5,5,0.5))
plt.xticks(np.arange(-0.5,5,0.5),np.arange(0,55,5))
plt.yticks(np.arange(-0.5,5,0.5),np.arange(0,55,5))
ax.set_xlabel('Distance (km)')
ax.set_ylabel('Distance (km)')
fig.savefig('isoline_1_hour.png',dpi=300)
```

###depth for second hour

```python
fig = plt.figure()
ax = fig.add_subplot(111)
cs = ax.contour(np.arange(5),np.arange(5),depth[1,:,:])  ###second hour
plt.clabel(cs,fmt='%1.1f')
#plt.xticks(np.arange(-0.5,5,0.5))
plt.xticks(np.arange(-0.5,5,0.5),np.arange(0,55,5))
plt.yticks(np.arange(-0.5,5,0.5),np.arange(0,55,5))
ax.set_xlabel('Distance (km)')
ax.set_ylabel('Distance (km)')
fig.savefig('isoline_2_hour.png',dpi=300)
```

###total depth for first storm

```python
depth_sum = np.zeros((1,5,5))
for j in np.arange(5):
    for k in np.arange(5):
        depth_sum[0,j,k] = np.sum(depth[0:3,j,k])
fig = plt.figure()
ax = fig.add_subplot(111)
cs = ax.contour(np.arange(5),np.arange(5),depth_sum[0,:,:])  ###second hour
plt.clabel(cs,fmt='%1.1f')
#plt.xticks(np.arange(-0.5,5,0.5))
plt.xticks(np.arange(-0.5,5,0.5),np.arange(0,55,5))
plt.yticks(np.arange(-0.5,5,0.5),np.arange(0,55,5))
ax.set_xlabel('Distance (km)')
ax.set_ylabel('Distance (km)')
fig.savefig('isoline_1_storm.png',dpi=300)
```
