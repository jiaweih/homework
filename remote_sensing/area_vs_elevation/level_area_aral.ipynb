```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import os
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
%matplotlib inline
```

```python
mpl.rcParams['xtick.labelsize'] = 16
mpl.rcParams['ytick.labelsize'] = 16
mpl.rcParams['lines.linewidth'] = 1
mpl.rcParams['axes.labelsize'] = 18
mpl.rcParams['legend.fontsize'] = 15
mpl.rcParams['figure.figsize'] = [10, 8]
mpl.rcParams['figure.dpi'] = 300
```

```python
df_mead = pd.DataFrame(np.genfromtxt('level.txt',skip_header=6,usecols=[0,1,2]),columns=['date','height','std'])
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(df_mead.date,df_mead.height,lw=3)
ax.set_ylabel('Height (m)')
fig.savefig('height_aral.png')
```

```python
df_surface = pd.DataFrame(np.genfromtxt('surface.txt',skip_header=4,usecols=[0,1]),columns=['date','surface'])
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(df_surface.date,df_surface.surface,lw=3)
ax.set_ylabel('Surface Area ($km^2$)')
ax.set_xlim([1992,2013])
majorLocator = MultipleLocator(2)
majorFormatter = FormatStrFormatter('%d')
ax.xaxis.set_major_locator(majorLocator)
ax.xaxis.set_major_formatter(majorFormatter)
fig.savefig('surface_aral.png')
```

```python
level = np.genfromtxt('level.txt',skip_header=6,usecols=[0,1])
surface = np.genfromtxt('surface.txt',skip_header=4,usecols=[0,1])
data = np.empty((len(surface),3))
for i in np.arange(len(surface)):
    for j in np.arange(len(level)):
        if level[j,0] == surface[i,0]:
            data[i,0] = level[j,1]
            data[i,1] = surface[i,1]
            data[i,2] = 0.001*0.5*(data[i,0] - data[0,0])*(data[i,1] + data[0,1])                       
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(data[:,0],data[:,1],'r*')
ax.plot(data[:,0],data[:,1],'b-')
ax.set_xlabel('Height (m)')
ax.set_ylabel('Surface Area ($km^2$)')
fig.savefig('heigth_surface.png')
```
