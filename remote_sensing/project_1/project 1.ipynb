```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from netCDF4 import Dataset,num2date
import os
from mpl_toolkits.basemap import Basemap,cm
import matplotlib as mpl
%matplotlib inline
```

##Plot multi-year mean

```python
os.chdir('/raid/jiawei/remote_sensing/project1/subset')
```

```python
def get_data(ifile,var_name='pcp',lat_name='latitude',lon_name='longitude'):
    nc = Dataset(ifile,'r')
    lat = nc.variables[lat_name][:]
    lon = nc.variables[lon_name][:]
    var = nc.variables[var_name][:]
    nc.close()
    return (var[0],lat,lon)
```

###Plot multi-year mean without mask

```python
var,lat,lon = get_data('timmean.daily.nc')
latmin = min(lat)
latmax = max(lat)
latmean = (latmin + latmax)/2.
lonmin = min(lon)
lonmax = max(lon)
lonmean = (lonmin + lonmax)/2.
```

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)
m = Basemap(projection = 'merc', lat_0=latmean,lon_0=lonmean,llcrnrlat=latmin,
            urcrnrlat=latmax,llcrnrlon=lonmin,urcrnrlon=lonmax,resolution='i')
m.drawcoastlines(linewidth=0.25)
xi,yi = m(*np.meshgrid(lon,lat))
norm = mpl.colors.Normalize(1,6)
cmap = cm.GMT_drywet
m.pcolormesh(xi,yi,var,cmap=cmap,norm=norm)
cax = [0.22,0.08,0.6,0.03]
cax = plt.axes(cax)
cbar = mpl.colorbar.ColorbarBase(cax,cmap=cmap,norm=norm,orientation='horizontal',extend='max')
cbar.ax.tick_params(labelsize=18)
cbar.set_label('Multi-year Mean Precipitation (mm/day)', fontsize = 15)
fig.savefig('mean daily mean.png',dpi=300)
```

##Plot monthly mean time series

```python
def get_data_ts(ifile,varname='pcp',timename='time'):
    nc = Dataset(ifile,'r')
    time = num2date(nc.variables[timename][:],
                    nc.variables[timename].units,
                    nc.variables[timename].calendar)
    df = pd.Series(nc.variables[varname][:,0,0],index=time)
    nc.close
    return df
```

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)
df = get_data_ts('fldmean.daily.nc')
df.plot()
ax.set_ylabel('Monthly Mean Precipitation (mm/day)',fontsize=15)
ax.tick_params(axis='both',which='major',labelsize=15)
fig.savefig('monthly mean.png',dpi=300)
```

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)
df = get_data_ts('fldmean.yearmean.nc')
df.plot()
ax.set_ylabel('Annual Mean Precipitation (mm/day)',fontsize=15)
ax.tick_params(axis='both',which='major',labelsize=15)
fig.savefig('yearmean_mean.png',dpi=300)
```

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)
df = get_data_ts('fldmean.yearmean.nc')
df = pd.DataFrame(df)
df.boxplot()
ax.set_ylabel('Monthly Mean Precipitation (mm/day)',fontsize=15)
ax.tick_params(axis='both',which='major',labelsize=15)
fig.savefig('yearmean mean.png',dpi=300)
```
