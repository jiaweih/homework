```python
def get_data(ifile,var_name='Prec',lat_name='latitude',lon_name='longitude'):
    if os.path.exists(ifile):
        nc = Dataset(ifile,'r')   
    else:
        sys.exit('No such file: {}'.format(ifile))
    lat = nc.variables[lat_name][:]
    lon = nc.variables[lon_name][:]
    var = nc.variables[var_name][:]
    nc.close()
    return (var[0],lat,lon)
```
