```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import os
import datetime as dt
import netCDF4
%matplotlib inline
```

```python
mpl.rcParams['xtick.labelsize'] = 16
mpl.rcParams['ytick.labelsize'] = 16
mpl.rcParams['lines.linewidth'] = 3
mpl.rcParams['axes.labelsize'] = 18
mpl.rcParams['legend.fontsize'] = 15
mpl.rcParams['figure.figsize'] = [10, 8]
```

```python
os.chdir('/raid/jiawei/remote_sensing/project3/')
```

```python
data = np.genfromtxt('american.flow.txt', skip_header=2, usecols=[3])
data.size
```

```python
start = dt.datetime(2011,10,1)
end = dt.datetime(2012,5,31)
dates = pd.date_range(start=start,end=end,freq='D')
df = pd.DataFrame(data, index = dates, columns=['flow'])
```

```python
nc = netCDF4.Dataset('american.nc','r')
data = nc.variables['snodp'][:,0,0]
```

```python
df['snow'] = data
df.head(10)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
df['flow'].plot(ax = ax1, color='b',label = 'streamflow')
ax1.set_ylabel('Streamflow (cfs)')
ax2 = ax1.twinx()
df['snow'].plot(ax = ax2,color='r',label = 'snow depth')
ax2.set_ylabel('Snow depth')
ax1.plot(0,0,'r',label = 'snow depth')
ax1.legend(loc = 'upper left')
fig.savefig('flow_snow.png',dpi=300)
```
