## Python scripts for plotting

### Importing python modules

```python
%matplotlib inline
import matplotlib.pylab as plt
import pandas as pd
from datetime import datetime
import numpy as np
```

```python
pillows_header = ['year', 'month', 'day', 'hr', 'minute', 'second', 'Tuolumne_SWE', 'Dana_SWE']
pillows_start = datetime(1996,10,1)
pillows_end = datetime(2007,10,1)
pillows_dates = pd.date_range(start = pillows_start, end = pillows_end, freq = 'D')
course_header = ['year', 'month', 'day', 'hr', 'minute', 'second', 'SWE']
```

### Loading data

```python
pillows_data = np.genfromtxt('Snow_Pillows', delimiter='\t')
course_data = np.genfromtxt('Dana_Snow_Course', delimiter='\t')
```

### Creating pandas dataframe

```python
course_df = pd.DataFrame(course_data, columns=course_header)
pillows_df = pd.DataFrame(pillows_data, index = pillows_dates, columns = pillows_header)
```

```python
yy = course_df['year']
mm = course_df['month']
dd = course_df['day']
course_dates = []
for i in range(len(yy)):
    course_dates.append(datetime(int(yy[i]),int(mm[i]),int(dd[i])))
```

### Using matplotlib to plot SWE

```python
fig1 = plt.figure(figsize = (12,8))
ax = fig1.add_subplot(111)
ax.plot(pillows_df.index,pillows_df['Dana_SWE'],label = 'Dana Pillow Measurements')
ax.plot(course_dates,course_df['SWE'],'k*',label = 'Dana Course Measurements')
ax.set_xlabel('Year')
ax.set_ylabel('SWE(mm)')
ax.legend(loc = 'upper center')
```

```python
fig2 = plt.figure(figsize = (12,8))
ax = fig2.add_subplot(111)
ax.plot(pillows_df.index,pillows_df['Dana_SWE'],label = 'Dana Pillow Measurements')
ax.plot(course_dates,course_df['SWE'],'k*',label = 'Dana Course Measurements')
ax.plot(pillows_df.index, pillows_df['Tuolumne_SWE'], label = 'Tuolumne Pillow Measurements')
ax.set_xlabel('Year')
ax.set_ylabel('SWE(mm)')
ax.legend(loc = 'upper center')
```
