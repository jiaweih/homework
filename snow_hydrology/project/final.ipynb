```python
%matplotlib inline
import pandas as pd
from datetime import datetime
import numpy as np
import matplotlib.pylab as plt
import matplotlib as mpl
```

```python
path = '/usr1/jiawei/homework/snowHydrology/project/longwave.txt'
header = ['temperature_C','relative_humidity','observed_longwave','observed_shortwave','Bird']
start = datetime(2013,2,1,0,30)
end = datetime(2013,2,16,23,30)
dates = pd.date_range(start=start,end=end,freq='30min')
sigma = 5.670373*1E-8
```

```python
def return_a_b(c1,c2):
    if c1>c2:
        print 'c2 should be greater than c2'
        return
    a = 1.0/(c2-c1)
    b = float(c2/(c2-c1))
    return a,b
```

```python
data = np.genfromtxt(path,delimiter='\t')
df = pd.DataFrame(data, index=dates, columns = header)
```

```python
algorithm_clear = ['Angstrom','Brunt', 'Brutsaert','Garratt','Idso_1', 'Keding',  'Satterlunk', 'Swinbank']
algorithm_cloud = ['Brutsaert','Jacobs','Kedding','Maykut','Sugita','Unsworth']
index = 'index_'
cover = 'cover_'
RMSD = 'RMSD_'
N = len(df)
len_clr = len(algorithm_clear)
len_cld = len(algorithm_cloud)
RMSD_clr = np.zeros(len_clr)
```

```python
df['temperature_K'] = df['temperature_C'] + 273.15
tem_K = df['temperature_K']
df['saturation_pressure'] = 0.1*np.exp(77.3450 + 0.0057*tem_K - 7235/tem_K)/tem_K**8.2
df['vapor_pressure'] = df['relative_humidity'] * df['saturation_pressure']/100
df['precipitable'] = 4650*df['vapor_pressure']/df['temperature_K']
df['coefficient'] = sigma*df['temperature_K']**4
```

```python
for clr in algorithm_clear:
    if clr == 'Angstrom':
        df[clr] = (0.83 - 0.18*10**(-0.067*df['vapor_pressure']))*df['coefficient']        
    elif clr == 'Brunt':
        df[clr] = (0.52 + 0.205*df['vapor_pressure']**0.5)*df['coefficient']
    elif clr == 'Brutsaert':
        df[clr] = 1.723*(df['vapor_pressure']/df['temperature_K'])**(1/7)*df['coefficient']
    elif clr == 'Garratt':
        df[clr] = (0.79 - 0.17*np.exp(-0.96*df['vapor_pressure']))*df['coefficient']
    elif clr == 'Idso_1':
        df[clr] = (1 - 0.261*np.exp(-0.00077*(df['temperature_K'] - 273.16)**2))*df['coefficient']
    elif clr == 'Keding':
        df[clr] = (0.92 - 0.7*10**(-1.2*df['vapor_pressure']))*df['coefficient']
    elif clr == 'Satterlunk':
        df[clr] = 1.08*(1 - np.exp(-(10 * df['vapor_pressure'])**df['temperature_K']/2016))*df['coefficient']
    elif clr == 'Swinbank':
        df[clr] = 5.31*10**(-13)*df['temperature_K']**6       
```

```python
fig1 = plt.figure(figsize=(15,10))
ax = fig1.add_axes([0.1,0.1,0.8,0.8])
ax.plot(dates, df['observed_longwave'],linestyle='-',label='Observed',linewidth=3)
for i,clr in enumerate(algorithm_clear):
    ax.plot(dates, df[clr],linestyle='-',label=clr)
ax.legend(prop={'size':12})
ax.set_title('Observed vs Estimated Longwave Radiation',fontsize=20)
ax.set_xlabel('Time',fontsize=18)
ax.set_ylabel('Longwave Radiation ($W/m^2$)',fontsize=18)
ax.tick_params(axis='both',which='major',labelsize=14)
ax.set_ylim(150,1000)
fig1.savefig('clear_longwave_radiation.png',dpi = 300)
```

```python
c1 = 0.4
c2 = 0.7
a,b = return_a_b(c1,c2)
for i,clr in enumerate(algorithm_clear):
    df[index + clr] = df['observed_shortwave']/df[clr]
    df[cover + clr] = b - a*df[index + clr]
    df[RMSD + clr] = (df['observed_longwave'] - df[clr])**2/N
    RMSD_clr[i] = np.sum(df[RMSD + clr])**0.5
    for cld in algorithm_cloud:
        if cld == 'Brutsaert':
            df[clr+'_'+cld] = (1+0.22*df[cover + clr])*df[clr]
        elif cld == 'Jacobs':
            df[clr+'_'+cld] = (1+0.26*df[cover + clr])*df[clr]
        elif cld == 'Kedding':
            df[clr+'_'+cld]=(1+0.153*df[cover + clr]**2.183)*df[clr]
        elif cld == 'Maykut':
            df[clr+'_'+cld]=(1+0.22*df[cover + clr]**2.75)*df[clr]
        elif cld == 'Sugita':
            df[clr+'_'+cld]=(1+0.0496*df[cover + clr]**2.45)*df[clr]
        elif cld == 'Unsworth':
            df[clr+'_'+cld]=((1-0.84*df[cover + clr])*df[clr]+0.84*df[cover + clr])
```

```python
RMSD_cld = np.zeros((len_clr,len_cld))
for i,clr in enumerate(algorithm_clear):
    for j,cld in enumerate(algorithm_cloud):
        df[RMSD + clr + '_' + cld] = (df['observed_longwave'] - df[clr+'_'+cld])**2/N
        RMSD_cld[i,j] = np.sum(df[RMSD + clr + '_' + cld])**0.5
```

```python
name_RMSD = ['RMSD']
df_RMSD = pd.DataFrame(RMSD_clr,index=algorithm_clear,columns=name_RMSD)
df_RMSD_cld = pd.DataFrame(RMSD_cld,index=algorithm_clear,columns=algorithm_cloud)
```
