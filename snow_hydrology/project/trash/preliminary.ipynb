```python
%matplotlib inline
import pandas as pd
from datetime import datetime
import numpy as np
import matplotlib.pylab as plt
import matplotlib as mpl
```

```python
path = '/usr1/jiawei/homework/snowHydrology/project/longwave.txt'
header = ['temperature_C','relative_humidity','observed_longwave']
start = datetime(2013,2,1,0,30)
end = datetime(2013,2,16,23,30)
dates = pd.date_range(start=start,end=end,freq='30min')
sigma = 5.670373*1E-8
```

```python
data = np.genfromtxt(path,delimiter='\t')
df = pd.DataFrame(data, index=dates, columns = header)
N = len(df)
```

```python
df['temperature_K'] = df['temperature_C'] + 273.15
tem_K = df['temperature_K']
df['saturation_pressure'] = 0.1*np.exp(77.3450 + 0.0057*tem_K - 7235/tem_K)/tem_K**8.2
df['vapor_pressure'] = df['relative_humidity'] * df['saturation_pressure']/100
df['precipitable'] = 4650*df['vapor_pressure']/df['temperature_K']
df['coefficient'] = sigma*df['temperature_K']**4
```

```python
df['Angstrom'] = (0.83 - 0.18*10**(-0.067*df['vapor_pressure']))*df['coefficient']

df['RMSD_Angstrom'] = (df['observed_longwave'] - df['Angstrom'])**2/N
RMSD_Angstrom = np.sum(df['RMSD_Angstrom'])**0.5
```

```python
df['Brunt'] = (0.52 + 0.205*df['vapor_pressure']**0.5)*df['coefficient']
df['RMSD_Brunt'] = ((df['observed_longwave'] - df['Brunt'])**2/N)
RMSD_Brunt = np.sum(df['RMSD_Brunt'])**0.5
```

```python
df['Brutsaert'] = 1.723*(df['vapor_pressure']/df['temperature_K'])**(1/7)*df['coefficient']
df['RMSD_Brutsaert'] = ((df['observed_longwave'] - df['Brutsaert'])**2/N)
RMSD_Brutsaert = np.sum(df['RMSD_Brutsaert'])**0.5
```

```python
df['Garratt'] = (0.79 - 0.17*np.exp(-0.96*df['vapor_pressure']))*df['coefficient']
df['RMSD_Garratt'] = ((df['observed_longwave'] - df['Garratt'])**2/N)
RMSD_Garratt = np.sum(df['RMSD_Garratt'])**0.5
```

```python
df['Idso_1'] = (1 - 0.261*np.exp(-0.00077*(df['temperature_K'] - 273.16)**2))*df['coefficient']
df['RMSD_Idso_1'] = ((df['observed_longwave'] - df['Idso_1'])**2/N)
RMSD_Idso_1 = np.sum(df['RMSD_Idso_1'])**0.5
```

```python
df['Idso_2'] = (0.7 + 5.95*1e-4*df['vapor_pressure']*np.exp(1500/df['temperature_K']))*df['coefficient']
df['RMSD_Idso_2'] = ((df['observed_longwave'] - df['Idso_2'])**2/N)
RMSD_Idso_2 = np.sum(df['RMSD_Idso_2'])**0.5
```

```python
df['Keding'] = (0.92 - 0.7*10**(-1.2*df['vapor_pressure']))*df['coefficient']
df['RMSD_Keding'] = ((df['observed_longwave'] - df['Keding'])**2/N)
RMSD_Keding = np.sum(df['RMSD_Keding'])**0.5
```

```python
df['Prata'] = (1 - (1 + df['precipitable'])*np.exp(-(1.2 + 3*df['precipitable'])**0.5))*df['coefficient']
df['RMSD_Prata'] = ((df['observed_longwave'] - df['Prata'])**2/N)
RMSD_Prata = np.sum(df['Prata'])**0.5
```

```python
df['Satterlunk'] = 1.08*(1 - np.exp(-(10 * df['vapor_pressure'])**df['temperature_K']/2016))*df['coefficient']
df['RMSD_Satterlunk'] = ((df['observed_longwave'] - df['Satterlunk'])**2/N)
RMSD_Satterlunk = np.sum(df['RMSD_Satterlunk'])**0.5
```

```python
df['Swinbank'] = 5.31*10**(-13)*df['temperature_K']**6
df['RMSD_Swinbank'] = ((df['observed_longwave'] - df['Swinbank'])**2/N)
RMSD_Swinbank = np.sum(df['RMSD_Swinbank'])**0.5
```

```python
df['Dilley'] = 59.38 + 113.7*(df['temperature_K']/273.16)**6 + 96.96*(df['precipitable']/25)**0.5
df['RMSD_Dilley'] = ((df['observed_longwave'] - df['Dilley'])**2/N)
RMSD_Dilley = np.sum(df['RMSD_Dilley'])**0.5
```

```python
RMSD = [RMSD_Angstrom,RMSD_Brunt,RMSD_Brutsaert,RMSD_Garratt,RMSD_Idso_1,RMSD_Idso_2,RMSD_Keding,
        RMSD_Prata,RMSD_Satterlunk,RMSD_Swinbank,RMSD_Dilley]
algorithm = ['Angstrom','Brunt', 'Brutsaert','Garratt','Idso_1', 'Idso_2', 'Keding', 'Prata', 'Satterlunk', 'Swinbank', 'Dilley']
name = ['RMSD']
```

```python
df_RMSD = pd.DataFrame(RMSD,index=algorithm,columns=name)
df_RMSD
```

```python
#dates = pd.date_range(start = start, end = end, freq = '30min')

fig1 = plt.figure(figsize=(20,15))
ax = fig1.add_axes([0.1,0.1,0.8,0.8])
ax.plot(dates, df['observed_longwave'],linestyle='-',label='Observed')
ax.plot(dates, df['Keding'],linestyle='-',label='Keding')
ax.plot(dates, df['Idso_1'],linestyle='-',label='Idso_1')
ax.plot(dates, df['Garratt'],linestyle='-',label='Garratt')
ax.plot(dates, df['Satterlunk'],linestyle='-',label='Satterlunk')
ax.plot(dates, df['Swinbank'],linestyle='-',label='Swinbank')
ax.legend(prop={'size':18})
ax.set_title('Observed vs Estimated Longwave Radiation',fontsize=20)
ax.set_xlabel('Time',fontsize=18)
ax.set_ylabel('Longwave Radiation',fontsize=18)
ax.tick_params(axis='both',which='major',labelsize=14)
```

```python
fig2 = plt.figure(figsize=(20,10))
ax = fig2.add_axes([0.1,0.1,0.8,0.8])
ax.plot(df_RMSD['RMSD'])
ax.set_xticks(np.arange(0,11))
ax.set_xticklabels(algorithm)
ax.set_xlabel('Algorithms', fontsize = 18)
ax.set_ylabel('RMSD', fontsize = 18)
fig2.savefig('RMSD.png')
ax.tick_params(axis='both',which='major',labelsize=14)
```
