```python
%matplotlib inline
import matplotlib.pylab as plt
import pandas as pd
from datetime import datetime
import numpy as np
import matplotlib
```

```python
SW_header = ['year', 'month', 'day', 'hr', 'minute', 'second', 'Tuolumne_SW', 'Dana_SW','NLDAS_SW']
SW_start = datetime(2003,1,1,1,0,0)
SW_end = datetime(2004,1,1,0,0,0)
SW_dates = pd.date_range(start = SW_start, end = SW_end, freq = 'h')
```

```python
SW_data = np.genfromtxt('radiation_data_2003_TUMDANNLDAS.txt', delimiter='\t')
```

```python
SW_df = pd.DataFrame(SW_data, index = SW_dates, columns = SW_header)
```

##1. Plot three timeseries

```python
fig0 = plt.figure(figsize = (12,8))
ax = fig0.add_subplot(111)
ax.plot(SW_dates,SW_df['Dana_SW'],color = 'blue',label = 'Dana Shortwave')
ax.plot(SW_dates,SW_df['Tuolumne_SW'],color = 'red',label = 'Tuolumne Shortwave')
ax.plot(SW_dates,SW_df['NLDAS_SW'],color = 'green',label = 'NLDAS Shortwave')
ax.legend(loc = 'best')
ax.set_xlabel('Time')
ax.set_ylabel('Shortwave ($W/m^2$)')
```

###One measurement against another in 2003

```python
fig1 = plt.figure(figsize = (12, 8))
ax = fig1.add_subplot(111)
ax.plot(SW_df['Dana_SW'],SW_df['Tuolumne_SW'],'k*',label = 'Dana Shortwave vs Tuolumne Shortwave')
ax.set_xlabel('Dana Shortwave ($W/m^2$)')
ax.set_ylabel('Tuolumne Shortwave ($W/m^2$)')
ax.legend(loc = 'upper left')

fig2 = plt.figure(figsize = (12, 8))
ax = fig2.add_subplot(111)
ax.plot(SW_df['Dana_SW'],SW_df['NLDAS_SW'],'k*',label = 'Dana Shortwave vs NLDAS Shortwave')
ax.set_xlabel('Dana Shortwave ($W/m^2$)')
ax.set_ylabel('NLDAS Shortwave ($W/m^2$)')
ax.legend(loc = 'best')

fig3 = plt.figure(figsize = (12, 8))
ax = fig3.add_subplot(111)
ax.plot(SW_df['Tuolumne_SW'],SW_df['NLDAS_SW'],'k*',label = 'Tuolumne Shortwave vs NLDAS Shortwave')
ax.set_xlabel('Tuolumne Shortwave ($W/m^2$)')
ax.set_ylabel('NLDAS Shortwave ($W/m^2$)')
ax.legend(loc = 'best')
```

>### From the timeseries graph above, measurements from Dana and Tuolumne were generally quite close to each other. Mesurement from NLDAS was generally lower than the other two. From the graph of comparison, at some points where the magnitude of shortwave radiation was somewhat low (0 - 200), records from NLDAS were much higher than Dana and Tuolumne records.

##2. Timeseries from March 3 to 5

```python
dates_1 = pd.date_range(start = datetime(2003,3,3,0,0,0), end = datetime(2003,3,6,0,0,0), freq = 'h')

fig4 = plt.figure(figsize = (12, 8))
ax = fig4.add_subplot(111)
ax.plot(dates_1,SW_df['Dana_SW'][dates_1],color = 'blue',label = 'Dana Shortwave')
ax.plot(dates_1,SW_df['Tuolumne_SW'][dates_1],color = 'red',label = 'Tuolumne Shortwave')
ax.plot(dates_1,SW_df['NLDAS_SW'][dates_1],color = 'green',label = 'NLDAS Shortwave')
ax.set_xlabel('Time')
ax.set_ylabel('Shortwave ($W/m^2$)')
ax.set_ylim([0,1000])
#ax.xaxis.set_major_locator(matplotlib.dates.DayLocator(bymonthday=None, interval=1, tz=None))
ax.xaxis.set_major_formatter(matplotlib.dates.DateFormatter('%H\n%b %d'))
#ax.xaxis.set_minor_locator(matplotlib.dates.HourLocator(byhour=None, interval=12, tz=None))
#ax.xaxis.set_minor_formatter(matplotlib.dates.DateFormatter('%H \n%b %d'))
ax.legend(loc = 'best')
```

>###The graph just shows there might be cloudy in Tuolumne on March 3 and cloudy in Dana on March 4. I don't think all of the measurements should be trusted. I think measurement from NLDAS were most reliable. As is shown in the graph above, the measurement from NLDAS was quite stable, while measurements from Dana and Tuolumne were not. Measurements from Dana and Tuolumne were significantly different from each other on March 3 and March 4. They cannot reflect the true radiation in a large scale.

##3. Timeseries from July 22 to 25

```python
dates_2 = pd.date_range(start = datetime(2003,7,22,0,0,0), end = datetime(2003,7,26,0,0,0), freq = 'h')

fig5 = plt.figure(figsize = (12, 8))
ax = fig5.add_subplot(111)
ax.plot(dates_2,SW_df['Dana_SW'][dates_2],color = 'blue',label = 'Dana Shortwave')
ax.plot(dates_2,SW_df['Tuolumne_SW'][dates_2],color = 'red',label = 'Tuolumne Shortwave')
ax.plot(dates_2,SW_df['NLDAS_SW'][dates_2],color = 'green',label = 'NLDAS Shortwave')
ax.set_xlabel('Time')
ax.set_ylabel('Shortwave ($W/m^2$)')
#ax.xaxis.set_major_locator(matplotlib.dates.DayLocator(bymonthday=None, interval=1, tz=None))
ax.xaxis.set_major_formatter(matplotlib.dates.DateFormatter('%H \n%b %d'))
ax.legend(loc = 'best')
```

>###The graph shows some perturbation of shortwave radiation during the daytime; in the morning it seemed fine, but in the afternoon there might be cloudy locally or it might be shading in Dana and Tuolumne measurements. I don't think all of the measurements should be trusted. I think measurement from NLDAS was most reliable. As is shown in the graph above, the measurement from NLDAS was quite stable, while measurements from Dana and Tuolumne were not. 

##4. Shading by surrounding terrain

```python
dates_3 = pd.date_range(start = datetime(2003,7,23,0,0,0), end = datetime(2003,7,24,0,0,0), freq = 'h')

fig5 = plt.figure(figsize = (12, 8))
ax = fig5.add_subplot(111)
ax.plot(dates_3,SW_df['Dana_SW'][dates_3],color = 'blue',label = 'Dana Shortwave')
ax.plot(dates_3,SW_df['Tuolumne_SW'][dates_3],color = 'red',label = 'Tuolumne Shortwave')
ax.plot(dates_3,SW_df['NLDAS_SW'][dates_3],color = 'green',label = 'NLDAS Shortwave')
ax.set_xlabel('Time')
ax.set_ylabel('Shortwave ($W/m^2$)')
#ax.xaxis.set_major_locator(matplotlib.dates.DayLocator(bymonthday=None, interval=1, tz=None))
ax.xaxis.set_major_formatter(matplotlib.dates.DateFormatter('%H \n%b %d'))
ax.legend(loc = 'best')
```

>###I think the graph above can be the evidence of local shading in the Tuolumne and Dana records. Since from 12:00 to 16:00, there were an apparent drop in shortwave radiation in both Dana and Tuolumne records while NLDAS records seemd normal. At other times during the day, the records seemed fine compared with each other.
